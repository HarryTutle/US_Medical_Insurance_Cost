US Medical Insurance costs project

Step one: import packages and data to work on.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('insurance.csv')
```

Step two: what do we have inside ?

```python
print(data.shape)
print(data.dtypes)
print(data.columns)

for val in data.columns:
    
    print(data[val].value_counts(dropna=False))
    
print(data.head(5))
```

So we have a dataset of 1338 rows and 7 variables. Some variables are objects
, maybe we will need to change them in numerical (sex, region, smoker).

```python
# We write functions to change data into numerical stuff

def sex(var):
    
    if var=='male':
        
        return 0
    
    elif var=='female':
        
        return 1
        
    
    
def smoke(var):
    
    if var=='no':
        
        return 0
    
    elif var=='yes':
        
        return 1
    
def region(var):
    
    if var=='southeast':
        
        return 0
    
    elif var=='southwest':
        
        return 1
    
    elif var=='northwest':
        
        return 2
    
    elif var=='northeast':
        
        return 3
    
    
    
# Now we change those variables:
print(data.head(20))

data['sex']=data['sex'].apply(lambda x: sex(x))
data['smoker']=data['smoker'].apply(lambda x: smoke(x))
data['region']=data['region'].apply(lambda x: region(x))

print(data.head(20))
print(data.dtypes)
```

We could simplify some variable and round them (charges, bmi)

```python
data['charges']=data['charges'].apply(lambda x: int(x))
data['bmi']= data['bmi'].apply(lambda x: int(x))

print(data.dtypes)
```

Now we will try to make a model able to predict charges depends on the others variables

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

target=data['charges']
dataset=data.drop(['charges'], axis=1)

X_train, X_test, y_train, y_test=train_test_split(dataset, target, random_state=243, test_size=0.2)

model=RandomForestRegressor()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

By changing the random_state, the score range is between 82 and 88 %.
